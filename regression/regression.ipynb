```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 6))

sqft = fig.add_subplot(121)
price = fig.add_subplot(122)

sqft.hist(df.sqft_living, bins=80,color="red")
sqft.set_xlabel('Ft^2')
sqft.set_title("SQFT")

price.hist(df.price, bins=80,color="purple")
price.set_xlabel("Price $")
price.set_title("Price")

plt.show()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

m = ols('price ~ sqft_living',df).fit()
print (m.summary())
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

m = ols('price ~ sqft_living + bedrooms + grade + condition',df).fit()
print (m.summary())
```

```python
import seaborn as sns
sns.jointplot(x="sqft_living", y="price", data=df, kind = 'reg',fit_reg= True, size = 10)
plt.show()
```
