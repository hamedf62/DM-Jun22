```python
resturants_feedbacks = []

for i, resturant in df.iterrows():
    if i<= 5000:
        # print(resturant["name"],resturant["id"])
        id = resturant["id"]
        feedbacks = requests.get(f"https://www.delino.com/restaurant/feedback/{id}/0/1000")
        feedbacks_json = feedbacks.json()
        if feedbacks_json:
            feedbacks_json_revised = [{**item, "resturant":resturant["name"]} for item in feedbacks_json]
            # print(resturant["name"])
            # print(feedbacks_json_revised)
            resturants_feedbacks = resturants_feedbacks + feedbacks_json_revised
    else:
        break

# resturants_feedbacks
rfdf = pd.DataFrame(resturants_feedbacks)
rfdf
```
