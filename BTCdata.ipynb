```python
btcCV = btc_candle[["Close","Volume"]].copy()
btcCV.head(2)
```

```python
btcCV.shape
```

```python
btcCV["change"] = (btcCV.Close - btcCV.Close.shift(1))
btcCV["change%"] = round(100*(btcCV.Close - btcCV.Close.shift(1))/btcCV.Close.shift(1),2)

btcCV
```
